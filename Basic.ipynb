### Operators

```python
#Arithmetic 

a = 5
b = 10

print(a+b)
print(a-b)
print(a/b)
print(a**b)  #5**3 = 5*5*5 = 125
print(b//a) #7/2 = 3.5 , 7//2 = 3  
print(b%a) #17/5 = 2(Reminder) , 17%5 = 2
```

```python
#Assignment

a = 5
a +=1   # a = a + 1
print(a)

b = 10
b -= 1   # b = b - 1
print(b)

c = 5
c *= 2   # c = c * 2
print(c)

d = 10
d **=2   # d = 10*10 = 100
print(d)

e = 10
e /= 2   # e = e/2 = 5
print(e)
```

```python
#Comparison

a = 10
print(10)

b = 10==5    # if both value is equal then ans is TRUE else False
print(b)

a = 34
b = 89
print(a>b)
print(a<b)
print(a>=b)
print(a<=b)
print(a!=b)
```

```python
#Logical

a = 10
b = 28

print(a==10 and b==28)   #Both values must be True 
print(a==10 and b==20)

print(a==10 or b==20)    #Atleast one value is True

print(not a==20)    #Opposite answer
```

```python
#Identity

x = 23
y = 42

print(x is y)    #True if both is similar
print(x is not y)   #True if both isn't similar
```

```python
#Membership

a = ["Tata","Mahindra","Suzuki"]
b = "Tata"
c = "BMW"

print(b in a)
print(c not in a) 
```

### Data Structures

List


```python
#List

list1 = [21,34,56,"List"]

print(type(list1))

print(list1[2])     #Indexing

list1[3] = "Array"   #Replace
print(list1)

list1.append("Maths")    # add
print(list1)

list1.insert(3,"Python")     # add using index
print(list1)

list1.remove("Array")       #Remove
print(list1)

list1.pop(0)     #Remove using index
print(list1)

del list1[1]     #Remove value using index
print(list1)

list1.clear()    #Clear whole list
print(list1)

print(len(list1))    #Length of list
```
