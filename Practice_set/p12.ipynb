```python
#1
data = { 
    "A": 45, 
    "B": 20, 
    "C": 70, 
    "D": 15 
}


min_key = 0
min_value = float("inf")

for key, value in data.items():
    if value < min_value:
        min_value = value
        min_key = key

print(min_key)
```
