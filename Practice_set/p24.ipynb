```python
#1 Print the second smallest number. 

numbers = [45, 12, 78, 9, 34, 18] 

first_small = float("inf")
second_small = float("inf")

for i in numbers:
    if first_small > i:
        second_small = first_small
        first_small = i

print(second_small)
```

```python
#2 Count how many numbers are divisible by both 3 and 5. 

numbers = [15, 22, 30, 45, 50, 60, 73] 

count = 0

for i in numbers:
    if i%3 == 0 and i%5 == 0:
        count += 1

print(count)
```

```python
#3 Create a new list containing only odd numbers greater than 20. 

numbers = [11, 24, 35, 42, 51, 18, 27, 10] 

new_list = []

for i in numbers:
    if i%2 != 0 and i >20:
        new_list.append(i)

print(new_list)
```

```python
#4 Find the product (multiplication) of all even numbers. 

numbers = [2, 5, 4, 7, 6] 

multi = 1

for i in numbers:
    if i%2 == 0:
        multi *= i

print(multi)
```

```python
#5 Find the student having the lowest marks. 

students = { 
    "A": 80, 
    "B": 65, 
    "C": 92, 
    "D": 58 
} 

lowest_marks = float("inf")
lowest_students = None

for key,value in students.items():
    if lowest_marks > value:
        lowest_students = key
        lowest_marks = value

print(lowest_students,":",lowest_marks)
```

```python
#6 Create a new dictionary containing only employees whose salary is between 40000 and 60000 (inclusive). 

employees = { 
    "Rahul": 55000, 
    "Amit": 32000, 
    "Neha": 48000, 
    "Jay": 61000, 
    "Riya": 40000 
} 

new = {}

for key,value in employees.items():
    if value >= 40000 and value <= 60000:
        new[key] = value

print(new)
```

```python
#7 Find the department having the lowest total salary. 

employees = [ 
    ["Rahul", "IT", 50000], 
    ["Amit", "HR", 30000], 
    ["Neha", "IT", 45000], 
    ["Jay", "Sales", 35000], 
    ["Riya", "HR", 40000] 
] 

department_salary = {}

for row in employees:
    department = row[1]
    salary = row[2]

    if department in department_salary:
        department_salary[department] += salary
    else:
        department_salary[department] = salary

lowest_salary = float("inf")
lowest_department = None

for key,value in department_salary.items():
    if lowest_salary > value:
        lowest_department = key
        lowest_salary = value

print(lowest_department,":",lowest_salary)
```

```python
#8 Count how many students scored above the average. 

students = { 
    "A": 80, 
    "B": 90, 
    "C": 70, 
    "D": 60, 
    "E": 95 
} 

avg = 0
count = 0

for key,value in students.items():
    avg += value
    count += 1

count1 = 0

for key,value in students.items():
    if value > (avg/count):
        count1 += 1

print(count1)
```

```python
#9 Find the employee with the second highest salary. 

employees = {
    "Rahul": 55000,
    "Amit": 32000,
    "Neha": 48000,
    "Jay": 61000,
    "Riya": 40000
}

first_highest = 0
second_highest = 0

first_employee = None
second_employee = None

for key, value in employees.items():

    if value > first_highest:
        second_highest = first_highest
        second_employee = first_employee

        first_highest = value
        first_employee = key

    elif value > second_highest and value < first_highest:
        second_highest = value
        second_employee = key

print(second_employee, ":", second_highest)
```

```python
#10 Find the category having the highest total sales. 

sales = [ 
["Laptop", "Electronics", 50000], 
["TV", "Electronics", 40000], 
["Chair", "Furniture", 15000], 
["Table", "Furniture", 18000], 
["Pen", "Stationery", 5000], 
["Notebook", "Stationery", 7000] 
] 

category_sales = {}

for row in sales:
    category = row[1]
    sale = row[2]

    if category in category_sales:
        category_sales[category] += sale
    else:
        category_sales[category] = sale

    
highest_category = None
highest_sales = 0

for key,value in category_sales.items():
    if highest_sales < value:
        highest_category = key
        highest_sales = value

print(highest_category,":",highest_sales)
```
