```python
#1  Find the third highest number.

numbers = [45, 78, 12, 99, 65, 87, 91] 

first = 0
second = 0
third = 0

for i in numbers:
    if i > first:
        third = second
        second = first 
        first = i

    elif i > second and i < first:
        third = second
        second = i

    elif i > third and i < second:
        third = i

print(third)
```

```python
#2 Count how many students scored below the average. 

students = { 
"A": 80, 
"B": 55, 
"C": 92, 
"D": 68, 
"E": 75 
}

total = 0

for key,value in students.items():
    total += value

count = 0

for key,value in students.items():
    if value < (total/len(students)):
        count += 1

print(count)
```

```python
#3 Create a new dictionary containing only products costing more than the average price. 

products = { 
"Laptop": 55000, 
"Mobile": 30000, 
"Tablet": 25000, 
"TV": 70000, 
"Monitor": 35000 
} 

total = 0

for key,value in products.items():
    total += value

new = {}

for key,value in products.items():
    if value > (total/len(products)):
        new[key] = value

print(new)
```

```python
#4 Find the department having the second highest total salary.

employees = [ 
["Rahul", "IT", 50000], 
["Amit", "HR", 30000], 
["Neha", "IT", 45000], 
["Jay", "Sales", 35000], 
["Riya", "HR", 40000], 
["Karan", "Sales", 50000] 
] 

department_salary = {}

for row in employees:
    department = row[1]
    salary = row[2]

    if department in department_salary:
        department_salary[department] += salary
    else:
        department_salary[department] = salary

first_department = None
second_department = None

first_salary = 0
second_salary = 0

for key,value in department_salary.items():
    if value > first_salary:
        second_salary = first_salary
        second_department = first_department

        first_department = key
        first_salary = value 

    elif value > second_salary:
        second_salary = value
        second_department = key

print(second_department,":",second_salary)
```

```python
#6 Create a new list containing numbers that are divisible by 2 and 3, but not divisible by 5. 

numbers = [12, 18, 20, 24, 30, 36, 40, 42, 60] 

new = []

for i in numbers:
    if i%2 == 0 and i%3 == 0 and i%5 != 0:
        new.append(i)

print(new)
```

```python
#10 Find the category having the highest average sales.

sales = [
    ["Laptop", "Electronics", 50000],
    ["TV", "Electronics", 40000],
    ["Camera", "Electronics", 30000],
    ["Chair", "Furniture", 15000],
    ["Table", "Furniture", 18000],
    ["Pen", "Stationery", 5000],
    ["Notebook", "Stationery", 7000]
]

category_total = {}
category_count = {}

# Total aur Count dono nikaalo
for row in sales:
    category = row[1]
    sale = row[2]

    if category in category_total:
        category_total[category] += sale
        category_count[category] += 1
    else:
        category_total[category] = sale
        category_count[category] = 1

highest_avg = 0
highest_category = None

# Average compare karo
for category in category_total:

    average = category_total[category] / category_count[category]

    if average > highest_avg:
        highest_avg = average
        highest_category = category

print(highest_category,":",highest_avg)
```
