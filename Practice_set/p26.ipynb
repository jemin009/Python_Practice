```python
#1 Find the third lowest number. 

numbers = [45, 12, 78, 9, 34, 18, 25] 


f_lowest = float("inf")
s_lowest = float("inf")
t_lowest = float("inf")

for i in numbers:
    if i < f_lowest:
        t_lowest = s_lowest
        s_lowest = f_lowest
        f_lowest = i

    elif i < s_lowest and i > f_lowest:
        t_lowest = s_lowest
        s_lowest = i

    elif i < t_lowest and i > s_lowest:
        t_lowest = i

print(t_lowest)
```

```python
#2 Count how many numbers are divisible by 4 or 6, but not by 8. 

numbers = [8, 12, 16, 18, 24, 32, 36, 40, 48] 

count = 0

for i in numbers:
    if (i%4 == 0 or i%6 == 0) and i%8 != 0:
        count += 1

print(count) 
```

```python
#3 Create a new dictionary containing only products whose price is less than the average price. 

products = { 
"Laptop": 55000, 
"Mobile": 30000, 
"Tablet": 25000, 
"TV": 70000, 
"Monitor": 35000 
}

total = 0
for key,value in products.items():
    total += value 

new = {}

for key,value in products.items():
    if value < (total/len(products)):
        new[key] = value

print(new)
```

```python
#4 Find the department having the highest average salary. 

employees = [ 
    ["Rahul", "IT", 50000], 
    ["Amit", "HR", 30000], 
    ["Neha", "IT", 45000], 
    ["Jay", "Sales", 35000], 
    ["Riya", "HR", 40000], 
    ["Karan", "Sales", 50000] 
]

dep_salary = {}
dep_count = {}

for row in employees:
    department = row[1]
    salary = row[2]

    if department in dep_salary:
        dep_salary[department] += salary
        dep_count[department] += 1
    else:
        dep_salary[department] = salary
        dep_count[department] = 1

high_avg = 0
high_dep = None

for department in dep_salary:
    average = dep_salary[department] / dep_count[department]

    if average > high_avg:
        high_avg = average
        high_dep = department

print(high_dep,":",high_avg)
```

```python
#8 Count how many students scored between 60 and 85 (inclusive). 

students = { 
"A": 80, 
"B": 55, 
"C": 92, 
"D": 68, 
"E": 75, 
"F": 60 
}

count = 0 

for key,value in students.items():
    if value >= 60 and value <= 85:
        count += 1

print(count)
```

```python
#9 Find the category having the lowest average sales. 

sales = [ 
["Laptop", "Electronics", 50000], 
["TV", "Electronics", 40000], 
["Camera", "Electronics", 30000], 
["Chair", "Furniture", 15000], 
["Table", "Furniture", 18000], 
["Pen", "Stationery", 5000], 
["Notebook", "Stationery", 7000] 
]

cat_total = {}
cat_count = {}

for row in sales:
    cat = row[1]
    sal = row[2]

    if cat in cat_total:
        cat_total[cat] += sal    
        cat_count[cat] += 1
    else:
        cat_total[cat] = sal    
        cat_count[cat] = 1

low_cat = None
low_sal = float("inf")

for cat in cat_total:
    average = cat_total[cat] / cat_count[cat]

    if average < low_sal:
        low_sal = average
        low_cat = cat

print(low_cat,":",low_sal)

```

```python
#10 Find the customer who spent the highest total amount. 

orders = [ 
["Rahul", 5000], 
["Amit", 3000], 
["Rahul", 4000], 
["Neha", 7000], 
["Amit", 6000], 
["Neha", 2000] 
]

cus_spent = {}

for row in orders:
    cus = row[0]
    spent = row[1]

    if cus in cus_spent:
        cus_spent[cus] += spent
    else:
        cus_spent[cus] = spent

high_cus = None
high_spent = 0

for key,value in cus_spent.items():
    if value > high_spent:
        high_cus = key
        high_spent = value

print(high_cus,":",high_spent)
```
