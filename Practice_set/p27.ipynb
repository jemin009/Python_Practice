```python
#3 Create a new dictionary containing only employees whose salary is greater than the department's average salary.

employees = [ 
    ["Rahul", "IT", 50000], 
    ["Neha", "IT", 45000], 
    ["Amit", "HR", 30000], 
    ["Riya", "HR", 40000], 
    ["Jay", "Sales", 35000], 
    ["Karan", "Sales", 50000] 
]

dep_total = {}
dep_count = {}

for row in employees:
    employee = row[0]
    department = row[1]
    salary = row[2]

    if department in dep_total:
        dep_total[department] += salary
        dep_count[department] += 1
    else:
        dep_total[department] = salary
        dep_count[department] = 1

new = {}

for row in employees:
    employee = row[0]
    department = row[1]
    salary = row[2]

    average = dep_total[department] / dep_count[department]

    if salary > average:
        new[employee] = salary

print(new)
```

```python
#5 Create a new list containing numbers that are greater than the average and divisible by 5. 

numbers = [15, 28, 35, 42, 50, 60, 18, 75] 

total = 0
count = 0

for i in numbers:
    total += i
    count+= 1

average = (total/count)

new = []

for i in numbers:
    if i%5 == 0 and i>average:
        new.append(i)

print(new)
```

```python
#8 Count how many departments have an average salary greater than 42000. 

employees = [ 
["Rahul", "IT", 50000], 
["Neha", "IT", 45000], 
["Amit", "HR", 30000], 
["Riya", "HR", 40000], 
["Jay", "Sales", 35000], 
["Karan", "Sales", 50000], 
["Pooja", "Finance", 60000], 
["Raj", "Finance", 25000] 
]

dep_total = {}
dep_count = {}

for row in employees:
    department = row[1]
    salary = row[2]

    if department in dep_total:
        dep_total[department] += salary
        dep_count[department] += 1
    else:
        dep_total[department] = salary
        dep_count[department] = 1

count = 0

for department in dep_total:
    average = dep_total[department] / dep_count[department]

    if average > 42000:
        count += 1

print(count)
```

```python
#10 Find the customer having the highest average order value. 

orders = [ 
["Rahul", 5000], 
["Rahul", 7000], 
["Amit", 3000], 
["Amit", 9000], 
["Neha", 8000], 
["Neha", 6000], 
["Jay", 10000] 
]

cus_total = {}
cus_count = {}

for row in orders:
    customer = row[0]
    value = row[1]

    if customer in cus_total:
        cus_total[customer] += value
        cus_count[customer] += 1
    else:
        cus_total[customer] = value
        cus_count[customer] = 1

high_value = 0
high_cus = None

for customer in cus_total:
    average = cus_total[customer] / cus_count[customer]

    if average > high_value:
        high_value = average
        high_cus = customer

print(high_cus,":",high_value)
```
