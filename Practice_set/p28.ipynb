```python
#10 Find the customer having the highest total order value, but consider only customers whose average order value is greater than 6000.

orders = [ 
    ["Rahul", 5000], 
    ["Rahul", 9000], 
    ["Amit", 3000], 
    ["Amit", 9000], 
    ["Neha", 8000], 
    ["Neha", 6000], 
    ["Jay", 10000], 
    ["Jay", 12000], 
    ["Riya", 7000] 
]

values_total = {}
cus_count = {}

for row in orders:
    customer = row[0]
    values = row[1]

    if customer in values_total:
        values_total[customer] += values
        cus_count[customer] += 1
    else:
        values_total[customer] = values
        cus_count[customer] = 1

high_value = 0
high_cus = None

for customer in values_total:
    total = values_total[customer]
    average = values_total[customer] / cus_count[customer]

    if total > high_value and average > 6000:
        high_value = total
        high_cus = customer

print(high_cus)
```
