```python
#1 Find the employee having the second highest average salary, but consider only employees whose total salary is greater than 100000. 

employees = [ 
    ["Rahul", 50000], 
    ["Rahul", 60000], 
    ["Amit", 40000], 
    ["Amit", 45000], 
    ["Neha", 70000], 
    ["Neha", 60000], 
    ["Jay", 30000], 
    ["Jay", 40000], 
    ["Riya", 80000] 
]

sal_total = {}
emp_count = {}

for row in employees:
    employee = row[0]
    salary = row[1]

    if employee in sal_total:
        sal_total[employee] += salary
        emp_count[employee] += 1
    else: 
        sal_total[employee] = salary
        emp_count[employee] = 1
        
sh_sal = 0
sh_emp = None

fh_sal = 0
fh_emp = None

for employee in sal_total:
    total = sal_total[employee]
    average = sal_total[employee] / emp_count[employee]

    if average > fh_sal and total > 100000:
        sh_sal = fh_sal
        sh_emp = fh_emp

        fh_sal = average
        fh_emp = employee

    elif total > 100000 and (average > sh_sal and average < fh_sal):
        sh_sal = average
        sh_emp = employee

print(sh_emp,":",sh_sal)
 
```

```python
#2 Count how many products have a price lower than the average price and are divisible by 4000 or 5000. 

products = { 
    "Laptop": 56000, 
    "Mobile": 32000, 
    "Tablet": 24000, 
    "TV": 72000, 
    "Monitor": 36000, 
    "Keyboard": 4000, 
    "Speaker": 16000, 
    "Mouse": 2000 
}

total = 0
count = 0

for key,value in products.items():
    total += value 
    count += 1

average = total/count
count = 0

for key,value in products.items():
    if (value % 4000 == 0 or value % 5000 == 0) and value < average:
        count += 1

print(count)
```

```python
#5 Create a new list containing numbers that are: 
'''
-> greater than the average  
-> divisible by 4  
-> not divisible by 6  
-> less than 80  
'''

numbers = [16, 24, 32, 40, 48, 60, 72, 84, 96] 

total = 0
count = 0

for i in numbers:
    total += i
    count += 1

average = total/count
new = []

for i in numbers:
    if i > average and i%4 == 0 and i%6 != 0 and i < 80:
        new.append(i)

print(new)
```
