```python
#1 Find the employee having the highest average salary, but consider only employees whose: 
'''
-> total salary is greater than 100000  
-> average salary is divisible by 5000  
'''

employees = [ 
    ["Rahul", 50000], 
    ["Rahul", 60000], 
    ["Amit", 45000], 
    ["Amit", 55000], 
    ["Neha", 70000], 
    ["Neha", 80000], 
    ["Jay", 40000], 
    ["Jay", 70000], 
    ["Riya", 60000], 
    ["Riya", 70000] 
]

sal_total = {}
emp_count = {}

for row in employees:
    employee = row[0]
    salary = row[1]

    if employee in sal_total:
        sal_total[employee] += salary
        emp_count[employee] += 1
    else:
        sal_total[employee] = salary
        emp_count[employee] = 1

high_sal = 0
high_cus = None

for employee in sal_total:
    total = sal_total[employee]
    average = sal_total[employee] / emp_count[employee]

    if average > high_sal and average % 5000 == 0 and total > 100000:
        high_sal = average 
        high_cus = employee

print(high_cus,":",high_sal)
```

```python
#2 Count how many products have: 
'''
-> price greater than the average price  
-> divisible by 4000 or 6000  
-> not divisible by 8000 
'''

products = { 
    "Laptop": 60000, 
    "Mobile": 36000, 
    "Tablet": 24000, 
    "TV": 72000, 
    "Monitor": 48000, 
    "Keyboard": 4000, 
    "Speaker": 18000, 
    "Mouse": 8000, 
    "Printer": 30000 
} 

total = 0
count = 0

for key,value in products.items():
    total += value
    count += 1

count2 = 0
for key,value in products.items():
    average = total/count
    if (value % 4000 == 0 or value % 6000 == 0) and value > average and value % 8000 != 0:
        count2 += 1

print(count2)
```

```python
#3 Create a new dictionary containing only employees whose: 
'''
-> salary is greater than their department average  
-> department total salary is greater than 85000  
-> salary is divisible by 5000 
'''

employees = [ 
    ["Rahul", "IT", 50000], 
    ["Neha", "IT", 60000], 
    ["Amit", "HR", 30000], 
    ["Riya", "HR", 50000], 
    ["Jay", "Sales", 35000], 
    ["Karan", "Sales", 55000], 
    ["Pooja", "Finance", 70000], 
    ["Raj", "Finance", 30000] 
]

sal_total = {}
dep_count = {}

for row in employees:
    department = row[1]
    salary = row[2]

    if department in sal_total:
        sal_total[department] += salary
        dep_count[department] += 1
    else:
        sal_total[department] = salary
        dep_count[department] = 1

new = {}

for row in employees:
    employee = row[0]
    department = row[1]
    salary = row[2]

    total = sal_total[department]
    average = sal_total[department] / dep_count[department]

    if salary > average and total > 85000 and salary % 5000 == 0:
        new[employee] = salary

print(new)
```
