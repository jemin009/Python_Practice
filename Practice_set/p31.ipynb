```python
#1 Find the employee having the second highest average salary, but consider only employees whose: 
'''
-> total salary is greater than 120000  
-> average salary is greater than the overall company average salary  
'''

employees = [ 
    ["Rahul", 60000], 
    ["Rahul", 70000], 
    ["Amit", 50000], 
    ["Amit", 60000], 
    ["Neha", 80000], 
    ["Neha", 70000], 
    ["Jay", 45000], 
    ["Jay", 85000], 
    ["Riya", 65000], 
    ["Riya", 70000] 
] 

sal_total = {}
emp_count = {}
o_total = 0
o_count = 0

for row in employees:
    employee = row[0]
    salary = row[1]

    if employee in sal_total:
        sal_total[employee] += salary
        emp_count[employee] += 1
        o_count += 1
        o_total += salary
    else:
        sal_total[employee] = salary
        emp_count[employee] = 1
        o_count += 1
        o_total += salary

o_average = o_total/o_count

fh_sal = 0
fh_emp = None

sh_sal = 0
sh_emp = None

for employee in sal_total:
    total = sal_total[employee]
    average = sal_total[employee] / emp_count[employee]

    if average > fh_sal and total > 120000 and average > o_average:
        sh_sal = fh_sal
        sh_emp = fh_emp

        fh_sal = average 
        fh_emp = employee

    elif (average > sh_sal and average < fh_sal) and total > 120000 and average > o_average:
        sh_sal = average
        sh_emp = employee

print(sh_emp,":",sh_sal)
```
