```python
#1 Create a new dictionary containing only those students whose total marks are greater than the overall class average total.

students = [ 
    ["Rahul", 78, 82, 91], 
    ["Neha", 88, 90, 85], 
    ["Amit", 65, 70, 72], 
    ["Riya", 95, 94, 96], 
    ["Jay", 80, 79, 81] 
]

mark_total = {}
stu_count = {}

for row in students:
    student = row[0]
    marks = row[1]+row[2]+row[3]

    if student in mark_total:
        mark_total[student] += marks
        stu_count[student] += 1
    else:
        mark_total[student] = marks
        stu_count[student] = 1

o_total = 0
for value in mark_total.values():
    o_total += value
o_average = o_total / len(mark_total)

new = {}
for key,value in mark_total.items():
    if value > o_average:
        new[key] = value

print(new)
```

```python
#2 Count how many products satisfy all conditions: 
'''
-> price is greater than the overall average price  
-> stock is less than 20  
-> price is divisible by 5000  
'''

products = { 
"Laptop": [75000, 10], 
"Mobile": [35000, 30], 
"Tablet": [25000, 15], 
"TV": [80000, 8], 
"Speaker": [15000, 12], 
"Monitor": [45000, 18] 
}

price_total = 0
count = 0
for key,value in products.items():
    price_total += value[0]
    count += 1
average = price_total/count

coun = 0
for key,value in products.items():
    if value[0] > average and value[1] < 20 and value[0] % 5000 == 0:
        coun += 1

print(coun)
```

```python
#4 Create a list containing only those numbers whose digit sum is even and the number is greater than the average.

numbers = [24,35,46,57,68,79,81,92,105] 

total = 0
count = 0
for i in numbers:
    total += i
    count += 1
average = total/count

new = []
for i in numbers:
    txt = str(i)

    digit_sum = 0
    for digit in txt:
        digit_sum += int(digit)

    if digit_sum % 2 == 0 and i > average:
        new.append(i)

print(new)
```

```python
#6 Create a dictionary whose key is department and value is average salary, but include only departments whose average salary is greater than the company average salary. 

employees = [ 
["Rahul","IT",60000], 
["Neha","IT",70000], 
["Amit","HR",45000], 
["Riya","HR",55000], 
["Jay","Sales",50000], 
["Karan","Sales",65000], 
["Pooja","Finance",85000], 
["Raj","Finance",35000] 
]

sal_total = {}
dep_count = {}
c_salary = 0
c_count = 0

for row in employees:
    department = row[1]
    salary = row[2]

    if department in sal_total:
        sal_total[department] += salary
        dep_count[department] += 1
    else:
        sal_total[department] = salary
        dep_count[department] = 1

    c_salary += salary
    c_count += 1

c_average = c_salary/c_count

new = {}
for department in sal_total:
    average = sal_total[department] / dep_count[department]

    if average > c_average:
        new[department] = average

print(new)
    
```

```python
#7 Find the product whose price is closest to the overall average price. 

products = { 
"Laptop":72000, 
"Mobile":36000, 
"Tablet":31000, 
"TV":85000, 
"Monitor":48000, 
"Keyboard":6000, 
"Speaker":18000 
}

total = 0
count = 0
for key,value in products.items():
    total += value
    count += 1
average = total/count


near_diff = float("inf")
near_price = 0
near_pro = None

for key, value in products.items():
    if abs(value - average) < near_diff:
        near_diff = abs(value - average)
        near_price = value
        near_pro = key

print(near_pro, ":", near_price)
```

```python
#8 Count how many students scored above the class average in at least 2 subjects.

students = [
    ["Rahul",70,80,90],
    ["Neha",90,95,85],
    ["Amit",60,65,70],
    ["Riya",95,90,100],
    ["Jay",75,85,80]
]

sub1_total = 0
sub2_total = 0
sub3_total = 0
count = 0

for row in students:
    sub1_total += row[1]
    sub2_total += row[2]
    sub3_total += row[3]
    count += 1

sub1_avg = sub1_total / count
sub2_avg = sub2_total / count
sub3_avg = sub3_total / count

student_count = 0

for row in students:

    above = 0

    if row[1] > sub1_avg:
        above += 1

    if row[2] > sub2_avg:
        above += 1

    if row[3] > sub3_avg:
        above += 1

    if above >= 2:
        student_count += 1

print(student_count)
```

```python
#9 Create a new list containing only those numbers that appear exactly once. 

numbers = [4,7,2,7,9,4,5,8,5,1,6]

freq = {}

for i in numbers:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1

new = []

for i in numbers:
    if freq[i] == 1:
        new.append(i)

print(new)
```

```python
#10 Find the customer whose average order value is closest to the overall average order value.

orders = [
    ["Rahul",8000],
    ["Rahul",12000],
    ["Amit",9000],
    ["Amit",11000],
    ["Neha",14000],
    ["Neha",16000],
    ["Jay",10000],
    ["Jay",12000],
    ["Riya",7000],
    ["Riya",9000]
]

value_total = {}
cus_count = {}
o_total = 0
o_count = 0

for row in orders:
    customer = row[0]
    value = row[1]

    if customer in value_total:
        value_total[customer] += value
        cus_count[customer] += 1
    else:
        value_total[customer] = value
        cus_count[customer] = 1

    o_total += value
    o_count += 1
o_average = o_total / o_count

close_diff = float("inf")
close_customer = None
close_average = 0

for customer in value_total:
    average = value_total[customer] / cus_count[customer]

    if abs(average - o_average) < close_diff:
        close_diff = abs(average - o_average)
        close_customer = customer
        close_average = average

print(close_customer, ":", close_average)
```
