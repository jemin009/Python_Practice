```python
#1 Create a dictionary containing only those students whose highest mark is greater than the overall highest subject average.

students = [
    ["Rahul",78,85,91],
    ["Neha",92,88,86],
    ["Amit",65,72,68],
    ["Riya",95,90,98],
    ["Jay",80,76,84]
]

# Highest mark of each student
high_mark = {}

for row in students:
    student = row[0]
    m1 = row[1]
    m2 = row[2]
    m3 = row[3]

    if m1 > m2 and m1 > m3:
        high_mark[student] = m1
    elif m2 > m1 and m2 > m3:
        high_mark[student] = m2
    else:
        high_mark[student] = m3

# Subject totals
m1_total = 0
m2_total = 0
m3_total = 0
count = 0

for row in students:
    m1_total += row[1]
    m2_total += row[2]
    m3_total += row[3]
    count += 1

# Subject averages
m1_average = m1_total / count
m2_average = m2_total / count
m3_average = m3_total / count

# Highest subject average
high_sub_avg = m1_average

if m2_average > high_sub_avg:
    high_sub_avg = m2_average

if m3_average > high_sub_avg:
    high_sub_avg = m3_average

# Final dictionary
new = {}

for student, mark in high_mark.items():
    if mark > high_sub_avg:
        new[student] = mark

print(new)
```
