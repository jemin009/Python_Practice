```python
#1. Create a dictionary containing only those students whose lowest mark is greater than the overall lowest subject average.

students = [ 
["Rahul",82,75,91], 
["Neha",90,88,86], 
["Amit",68,72,65], 
["Riya",95,94,98], 
["Jay",78,80,84] 
]

stu_low = {}

for row in students:
    student = row[0]
    m1 = row[1]
    m2 = row[2]
    m3 = row[3]

    if m1 < m2 and m1 < m3:
        stu_low[student] = m1
    elif m2 < m1 and m2 < m3:
        stu_low[student] = m2
    else:
        stu_low[student] = m3

m1_total = 0
m2_total = 0
m3_total = 0
count = 0

for row in students:
    student = row[0]
    m1 = row[1]
    m2 = row[2]
    m3 = row[3]

    m1_total += m1
    m2_total += m2
    m3_total += m3
    count += 1
        
m1_avg = m1_total/count
m2_avg = m2_total/count
m3_avg = m3_total/count

low_avg = 0

if m1_avg < m2_avg and m1_avg < m3_avg:
    low_avg = m1_avg
elif m2_avg < m1_avg and m2_avg < m3_avg:
    low_avg = m2_avg
else:
    low_avg = m3_avg

new = {}

for student,mark in stu_low.items():
    if mark > low_avg:
        new[student] = mark

print(new)
```

```python
#2 Count how many products have a stock value (price × quantity) that is less than the overall average stock value. 

products = { 
"Laptop":[70000,6], 
"Mobile":[40000,15], 
"Tablet":[30000,8], 
"TV":[90000,4], 
"Monitor":[50000,10], 
"Speaker":[20000,20] 
}

pro_value = {}
total_value = 0
count = 0

for key,value in products.items():
    pro_value[key] = value[0]*value[1]
    total_value += value[0]*value[1]
    count += 1
average = total_value/count

coun = 0
for product,value in pro_value.items():
    if value < average:
        coun += 1

print(coun)
```

```python
#3 Find the department having the highest employee count below the overall average employees per department. 

employees = [ 
["Rahul","IT"], 
["Neha","IT"], 
["Amit","HR"], 
["Riya","HR"], 
["Jay","Sales"], 
["Karan","Finance"], 
["Pooja","Finance"], 
["Raj","Finance"], 
["Vijay","Support"] 
]

dep_count = {}
total_dep = 0
count = 0

for row in employees:
    department = row[1]

    if department in dep_count:
        dep_count[department] += 1
    else:
        dep_count[department] = 1
        total_dep += 1
    count += 1
average = count/total_dep

high_count = 0
high_dep = None

for department,emp in dep_count.items():
    if emp < average and emp > high_count:
        high_count = emp
        high_dep = department

print(high_dep, ":", high_count)
```

```python
#6 Create a dictionary whose key is department and value is total salary, but include only departments whose total salary is below the company average department total.


employees = [ 
["Rahul","IT",60000], 
["Neha","IT",70000], 
["Amit","HR",40000], 
["Riya","HR",50000], 
["Jay","Sales",55000], 
["Karan","Sales",65000], 
["Pooja","Finance",80000], 
["Raj","Finance",30000] 
] 

dep_sal = {}
total_sal = 0
count = 0

for row in employees:
    department = row[1]
    salary = row[2]

    if department in dep_sal:
        dep_sal[department] += salary
    else:
        dep_sal[department] = salary
        count += 1
    total_sal += salary
average = total_sal/count
    
new = {}
for department,salary in dep_sal.items():
    total = dep_sal[department]

    if total < average:
        new[department] = total

print(new)
```

```python
#7 Find the product whose stock value (price × quantity) is farthest from the overall average stock value. 

products = { 
"Laptop":[70000,8], 
"Mobile":[35000,18], 
"Tablet":[25000,12], 
"TV":[80000,5], 
"Monitor":[45000,10], 
"Speaker":[15000,30] 
}

pro_value = {}
total_value = 0
count = 0

for key,value in products.items():
    pro_value[key] = value[0]*value[1]
    total_value += value[0]*value[1]
    count += 1
average = total_value/count

far_value = 0
far_pro = None

for product,value in pro_value.items():
    if abs(value - average) > far_value:
        far_value = abs(value - average)
        far_pro = product

print(far_pro,":",far_value)

```

```python
#8 Count how many students scored below the class average in exactly one subject. 

students = [
["Rahul",82,78,90],
["Neha",95,91,84],
["Amit",65,72,68],
["Riya",96,94,98],
["Jay",80,85,79]
]

# Subject totals
s1_total = 0
s2_total = 0
s3_total = 0
count = 0

for row in students:
    s1_total += row[1]
    s2_total += row[2]
    s3_total += row[3]
    count += 1

s1_avg = s1_total / count
s2_avg = s2_total / count
s3_avg = s3_total / count

# Count students below average in exactly one subject
student_count = 0

for row in students:
    below = 0

    if row[1] < s1_avg:
        below += 1

    if row[2] < s2_avg:
        below += 1

    if row[3] < s3_avg:
        below += 1

    if below == 1:
        student_count += 1

print(student_count)
```

```python
#9 Create a new list containing only those numbers whose digit sum is equal to the average digit sum of all numbers. 

numbers = [24,33,42,51,60,15,78,87,96] 

total = 0
count = 0

for i in numbers:
    txt = str(i)

    digit_sum = 0
    for digit in txt:
        digit_sum += int(digit)
    total += digit_sum
    count += 1
average = total / count

new = []

for i in numbers:
    txt = str(i)

    digit_sum = 0
    for digit in txt:
        digit_sum += int(digit)

    if digit_sum == average:
        new.append(i)

print(new)
```

```python
#10 Find the customer whose total order value is farthest from the overall average total order value. 

orders = [ 
["Rahul",9000], 
["Rahul",11000], 
["Amit",8000], 
["Amit",10000], 
["Neha",15000], 
["Neha",17000], 
["Jay",12000], 
["Jay",13000], 
["Riya",7000], 
["Riya",9000] 
]

cus_value = {}
total_value = 0
count = 0

for row in orders:
    customer = row[0]
    values = row[1]

    if customer in cus_value:
        cus_value[customer] += values
    else:
        cus_value[customer] = values
        count += 1
    total_value += values
average = total_value/count

far_value = 0
far_cust = None

for customer,values in cus_value.items():
    if abs(values - average) > far_value:
        far_value = abs(values - average)
        far_cust = customer

print(far_cust,":",far_value)
```
