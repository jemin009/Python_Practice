```python
#1 Create a dictionary containing only those students whose middle mark (after sorting their three marks) is greater than the overall average of all middle marks.

students = [ 
["Rahul",81,75,92], 
["Neha",90,87,84], 
["Amit",65,70,68], 
["Riya",95,99,94], 
["Jay",78,80,82] 
]

stu_mid = {}

# Find middle mark of each student
for row in students:
    student = row[0]

    marks = [row[1], row[2], row[3]]
    marks.sort()

    stu_mid[student] = marks[1]

# Find average of all middle marks
total = 0
count = 0

for mark in stu_mid.values():
    total += mark
    count += 1

average = total / count

# Create new dictionary
new = {}

for student, mark in stu_mid.items():
    if mark > average:
        new[student] = mark

print(new)
```

```python
#2 Count how many products have a price per unit stock (price ÷ quantity) greater than the overall average price per unit stock. .

products = { 
"Laptop":[72000,8], 
"Mobile":[36000,12], 
"Tablet":[30000,6], 
"TV":[84000,7], 
"Monitor":[48000,10], 
"Speaker":[18000,5] 
} 

pro_price = {}
total_price = 0
total_count = 0

for key,value in products.items():
    pro_price[key] = value[0]/value[1]
    total_price += value[0]/value[1]
    total_count += 1
average = total_price/total_count

count = 0
for product,price in pro_price.items():
    if price > average:
        count += 1

print(count)
```

```python
#6 Create a dictionary whose key is department and value is employee count, but include only departments whose employee count is greater than the overall average employee count per department.

employees = [ 
["Rahul","IT"], 
["Neha","IT"], 
["Amit","HR"], 
["Riya","HR"], 
["Jay","Sales"], 
["Karan","Sales"], 
["Pooja","Finance"], 
["Raj","Finance"], 
["Vijay","Finance"] 
]

dep_emp = {}
total_count = 0
total_dep = 0

for row in employees:
    department = row[1]

    if department in dep_emp:
        dep_emp[department] += 1
    else:
        dep_emp[department] = 1
        total_dep += 1
    total_count += 1
average = total_count/total_dep

new = {}
for department,employee in dep_emp.items():
    if employee > average:
        new[department] = employee

print(new)
```

```python
#7 Find the product whose quantity is closest to the overall average quantity. 

products = { 
"Laptop":[70000,8], 
"Mobile":[35000,15], 
"Tablet":[25000,12], 
"TV":[80000,5], 
"Monitor":[45000,10], 
"Speaker":[15000,18] 
}
total_qty = 0
total_count = 0

for key,value in products.items():
    total_qty += value[1]
    total_count += 1
average = total_qty/total_count

close_value = float("inf")
close_pro = None
for key,value in products.items():
    if abs(value[1] - average) < close_value:
        close_value = abs(value[1] - average)
        close_pro = key
 
print(close_pro,":",close_value)
```

```python
#8 Count how many students scored above the class average in exactly two subjects. 

students = [ 
["Rahul",82,76,91], 
["Neha",90,88,84], 
["Amit",68,70,65], 
["Riya",95,97,96], 
["Jay",80,79,83] 
]

stu_mark = {}
m1_total = 0
m2_total = 0
m3_total = 0
total_count = 0

for row in students:
    student = row[0]
    m1 = row[1]
    m2 = row[2]
    m3 = row[3]

    m1_total += m1
    m2_total += m2
    m3_total += m3
    total_count += 1

m1_avg = m1_total/total_count
m2_avg = m2_total/total_count
m3_avg = m3_total/total_count

count = 0

for row in students:
    above = 0

    if row[1] > m1_avg:
        above += 1

    if row[2] > m2_avg:
        above += 1

    if row[3] > m3_avg:
        above += 1

    if above == 2:
        count += 1

print(count)
```

```python
#9 Create a new list containing only those numbers whose product of digits is greater than the average product of digits of all numbers. 

numbers = [12,23,34,45,56,67,78,89,91] 

total = 0
count = 0
for i in numbers:
    txt = str(i)
    total += int(txt[0]) * int(txt[1])
    count += 1
average = total/count

new = []
for i in numbers:
    txt = str(i)

    if int(txt[0]) * int(txt[1]) > average:
        new.append(i)

print(new)

    
```
