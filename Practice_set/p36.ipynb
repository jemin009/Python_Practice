```python
#1 Create a dictionary containing only those students whose highest mark is less than the overall highest subject average. 

students = [ 
["Rahul",72,81,76], 
["Neha",88,91,85], 
["Amit",64,69,67], 
["Riya",93,89,95], 
["Jay",79,77,82] 
] 

stu_mark = {}

for row in students:
    student = row[0]
    m1 = row[1]
    m2 = row[2]
    m3 = row[3]

    if m1 > m2 and m1 > m3:
        stu_mark[student] = m1
    elif m2 > m1 and m2 > m3:
        stu_mark[student] = m2
    else:
        stu_mark[student] = m3

m1_total = 0
m2_total = 0
m3_total = 0
total_count = 0

for row in students:
    student = row[0]
    m1 = row[1]
    m2 = row[2]
    m3 = row[3]

    m1_total += m1
    m2_total += m2
    m3_total += m3
    total_count += 1

m1_avg = m1_total/total_count
m2_avg = m2_total/total_count
m3_avg = m3_total/total_count

high_avg = 0
if m1_avg > m2_avg and m1_avg > m3_avg:
    high_avg = m1_avg
elif m2_avg > m1_avg and m2_avg > m3_avg:
    high_avg = m2_avg
else:
    high_avg = m3_avg

new = {}
for student,mark in stu_mark.items():
    if mark < high_avg:
        new[student] = mark

print(new)
```

```python
#2 Count how many products have a stock value (price × quantity) less than the average stock value. 

products = { 
"Laptop":[68000,7], 
"Mobile":[42000,14], 
"Tablet":[28000,9], 
"TV":[76000,4], 
"Monitor":[52000,11], 
"Speaker":[16000,22] 
}

pro_values = {}
total_values = 0
total_count = 0

for product,value in products.items():
    pro_values[product] = value[0]*value[1]
    total_values += value[0]*value[1]
    total_count += 1
average = total_values/total_count

count = 0
for product,values in pro_values.items():
    if values < average:
        count += 1

print(count)
```

```python
#3 Find the department having the highest total salary below the overall average department total salary. 

employees = [ 
["Rahul","IT",58000], 
["Neha","IT",62000], 
["Amit","HR",45000], 
["Riya","HR",48000], 
["Jay","Sales",75000], 
["Karan","Sales",50000], 
["Pooja","Finance",95000], 
["Raj","Finance",35000] 
]

dep_sal = {}
sal_total = 0
dep_count = 0

for row in employees:
    department = row[1]
    salary = row[2]

    if department in dep_sal:
        dep_sal[department] += salary 
    else:
        dep_sal[department] = salary 
        dep_count += 1
    sal_total += salary
average = sal_total/dep_count

new = {}
for department,salary in dep_sal.items():
    dep_total = dep_sal[department]

    if dep_total < average:
        new[department] = salary

print(new)
```

```python
#6 Create a dictionary whose key is department and value is average salary, but include only departments whose average salary is less than the company average salary. 

employees = [ 
["Rahul","IT",65000], 
["Neha","IT",75000], 
["Amit","HR",42000], 
["Riya","HR",52000], 
["Jay","Sales",60000], 
["Karan","Sales",70000], 
["Pooja","Finance",85000], 
["Raj","Finance",45000] 
]

dep_sal = {}
dep_count = {}
total_sal = 0
total_count = 0

for row in employees:
    department = row[1]
    salary = row[2]

    if department in dep_sal:
        dep_sal[department] += salary
        dep_count[department] += 1
    else:
        dep_sal[department] = salary
        dep_count[department] = 1
    total_sal += salary
    total_count += 1
c_average = total_sal/total_count

new = {}
for department in dep_sal:
    average = dep_sal[department]/dep_count[department]

    if average < c_average:
        new[department] = average

print(new)
```

```python
#7 Find the product whose price is farthest from the overall average price. 

products = { 
"Laptop":[72000,6], 
"Mobile":[38000,18], 
"Tablet":[26000,10], 
"TV":[92000,5], 
"Monitor":[50000,12], 
"Speaker":[18000,20] 
} 

pro_price = {}
total_price = 0
total_count = 0

for key,value in products.items():
    pro_price[key] = value[0]
    total_price += value[0]
    total_count += 1
average = total_price/total_count

far_price = 0
far_pro = 0

for product,price in pro_price.items():
    if abs(price - average) > far_price:
        far_price = abs(price - average)
        far_pro = product

print(far_pro,":",far_price)
```

```python
#8 Count how many students scored below the class average in exactly one subject. 

students = [ 
["Rahul",84,79,91], 
["Neha",91,93,88], 
["Amit",66,71,68], 
["Riya",97,95,99], 
["Jay",81,78,85] 
]

stu_mark = {}
m1_total = 0
m2_total = 0
m3_total = 0
total_count = 0

for row in students:
    student = row[0]
    m1 = row[1]
    m2 = row[2]
    m3 = row[3]

    m1_total += m1
    m2_total += m2
    m3_total += m3
    total_count += 1

m1_avg = m1_total/total_count
m2_avg = m2_total/total_count
m3_avg = m3_total/total_count

count = 0
for row in students:
    student = row[0]
    m1 = row[1]
    m2 = row[2]
    m3 = row[3]

    below = 0
    if m1 < m1_avg:
        below += 1

    if m2 < m2_avg:
        below += 1

    if m3 < m3_avg:
        below += 1
    
    if below == 1:
        count += 1

print(count)
```

```python
#9 Create a new list containing only those numbers whose difference between first and last digit is greater than the average difference of all numbers. 

numbers = [18,27,39,45,56,64,73,82,91] 

num_dif = {}
dif_total = 0
dif_count = 0

new = []
for i in numbers:
    txt = str(i)
    dif = max((int(txt[0])),(int(txt[1]))) - min((int(txt[0])),(int(txt[1])))

    num_dif[i] = dif
    dif_total += dif
    dif_count += 1
    average = dif_total/dif_count

for num, dif in num_dif.items():
    if dif > average:
        new.append(num)

print(new)
```

```python
#10 Find the customer whose total order value is closest to the overall average total order value.

orders = [ 
["Rahul",10000], 
["Rahul",12000], 
["Amit",9000], 
["Amit",11000], 
["Neha",17000], 
["Neha",15000], 
["Jay",13000], 
["Jay",14000], 
["Riya",8000], 
["Riya",9000] 
]

cus_value = {}
total_values = 0
total_count = 0

for row in orders:
    customer = row[0]
    values = row[1]

    if customer in cus_value:
        cus_value[customer] += values
    else:
        cus_value[customer] = values
        total_count += 1
    total_values += values
average = total_values/total_count

close_value = float("inf")
close_cust = None

for customer,values in cus_value.items():
    if abs(values - average) < close_value:
        close_value = abs(values - average)
        close_cust = customer

print(close_cust,":",close_value)
```
