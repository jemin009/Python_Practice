```python
#1 Create a dictionary containing only those students whose lowest mark is greater than the overall average of the lowest marks. 

students = [ 
["Rahul",74,81,79], 
["Neha",92,88,85], 
["Amit",63,69,66], 
["Riya",96,91,98], 
["Jay",77,82,80] 
]

stu_mark = {}
mark_total = 0
stu_count = 0

for row in students:
    student = row[0]
    m1 = row[1]
    m2 = row[2]
    m3 = row[3]

    if m1 < m2 and m1 < m3:
        stu_mark[student] = m1
        mark_total += m1
    elif m2 < m1 and m2 < m3:
        stu_mark[student] = m2
        mark_total += m2
    else:
        stu_mark[student] = m3
        mark_total += m3
    stu_count += 1
average = mark_total/stu_count

new = {}
for student,mark in stu_mark.items():
    if mark > average:
        new[student] = mark

print(new)
```

```python
#2 Count how many products have a quantity greater than the overall average quantity. 

products = { 
"Laptop":[75000,9], 
"Mobile":[38000,16], 
"Tablet":[27000,7], 
"TV":[88000,5], 
"Monitor":[49000,11], 
"Speaker":[17000,18] 
} 

pro_qty = {}
total_qty = 0
total_count = 0
for key,value in products.items():
    pro_qty[key] = value[1]
    total_qty += value[1]
    total_count += 1
average = total_qty/total_count

count = 0
for product,qty in pro_qty.items():
    if qty > average:
        count += 1

print(count)
```

```python
#3 Find the department having the lowest average salary above the company average salary.

employees = [ 
["Rahul","IT",70000], 
["Neha","IT",76000], 
["Amit","HR",45000], 
["Riya","HR",55000], 
["Jay","Sales",68000], 
["Karan","Sales",72000], 
["Pooja","Finance",90000], 
["Raj","Finance",50000] 
]

dep_sal = {}
dep_emp = {}
total_sal = 0
total_count = 0

for row in employees:
    department = row[1]
    salary = row[2]

    if department in dep_sal:
        dep_sal[department] += salary
        dep_emp[department] += 1
    else:
        dep_sal[department] = salary
        dep_emp[department] = 1
    
    total_sal += salary
    total_count += 1
c_average = total_sal/total_count

low_salary = float("inf")
low_dep = None

for department,salary in dep_sal.items():
    d_average = dep_sal[department] / dep_emp[department]

    if d_average > c_average and d_average < low_salary:
        low_salary = d_average
        low_dep = department

print(low_dep,":",low_salary)
```

```python
#6 Create a dictionary whose key is department and value is total salary, but include only departments whose total salary is greater than the company average department total salary. 

employees = [ 
["Rahul","IT",68000], 
["Neha","IT",72000], 
["Amit","HR",43000], 
["Riya","HR",47000], 
["Jay","Sales",62000], 
["Karan","Sales",78000], 
["Pooja","Finance",88000], 
["Raj","Finance",52000] 
]

dep_sal = {}
total_sal = 0
total_count = 0

for row in employees:
    department = row[1]
    salary = row[2]

    if department in dep_sal:
        dep_sal[department] += salary
    else:
        dep_sal[department] = salary
        total_count += 1
    total_sal += salary
d_average = total_sal/total_count

new = {}
for department,salary in dep_sal.items():
    if salary > d_average:
        new[department] = salary

print(new)
```

```python
#7 Find the product whose quantity is farthest from the overall average quantity. 

products = { 
"Laptop":[72000,8], 
"Mobile":[36000,17], 
"Tablet":[24000,9], 
"TV":[90000,4], 
"Monitor":[48000,12], 
"Speaker":[18000,21] 
} 

pro_qty = {}
total_qty = 0
total_count = 0

for key,value in products.items():
    pro_qty[key] = value[1]
    total_qty += value[1]
    total_count += 1
average = total_qty/total_count

far_value = 0
far_product = None

for product,qty in pro_qty.items():
    if abs(qty - average) > far_value:
        far_value = abs(qty - average)
        far_product = product

print(far_product,":",far_value)
```

```python
#8 Count how many students scored above the class average in exactly one subject. 

students = [ 
["Rahul",85,78,92], 
["Neha",93,91,87], 
["Amit",67,72,69], 
["Riya",98,96,99], 
["Jay",82,80,84] 
] 

stu_mark = {}
m1_total = 0
m2_total = 0
m3_total = 0
total_count = 0

for row in students:
    student = row[0]
    m1 = row[1]
    m2 = row[2]
    m3 = row[3]

    m1_total += m1
    m2_total += m2
    m3_total += m3
    total_count += 1

m1_avg = m1_total/total_count
m2_avg = m2_total/total_count
m3_avg = m3_total/total_count

count = 0
for row in students:
    student = row[0]
    m1 = row[1]
    m2 = row[2]
    m3 = row[3]
    above = 0

    if m1 > m1_avg:
        above += 1

    if m2 > m2_avg:
        above += 1

    if m3 > m3_avg:
        above += 1
    
    if above == 1:
        count += 1

print(count)
```

```python
#9 Create a new list containing only those numbers whose product of first and last digit is less than the average product of all numbers. 

numbers = [13,24,35,46,57,68,79,81,92] 


total = 0
count = 0

for i in numbers:
    txt = str(i)
    product = int(txt[0]) * int(txt[-1])
    total += product
    count += 1
average = total/count

new = []
for i in numbers:
    txt = str(i)
    product = int(txt[0]) * int(txt[-1])

    if product < average:
        new.append(i)

print(new)
```

```python
#10 Find the customer whose average order value is closest to the overall average of customer average order values.

orders = [
["Rahul",11000],
["Rahul",13000],
["Amit",9000],
["Amit",10000],
["Neha",16000],
["Neha",18000],
["Jay",12000],
["Jay",14000],
["Riya",8000],
["Riya",9000]
]

cus_value = {}
cus_count = {}

# Total value and order count of each customer
for row in orders:
    customer = row[0]
    value = row[1]

    if customer in cus_value:
        cus_value[customer] += value
        cus_count[customer] += 1
    else:
        cus_value[customer] = value
        cus_count[customer] = 1

# Overall average of customer averages
total_avg = 0
count = 0

for customer in cus_value:
    avg = cus_value[customer] / cus_count[customer]
    total_avg += avg
    count += 1

o_average = total_avg / count

# Find closest customer
close_value = float("inf")
close_customer = None

for customer in cus_value:
    avg = cus_value[customer] / cus_count[customer]

    if abs(avg - o_average) < close_value:
        close_value = abs(avg - o_average)
        close_customer = customer

print(close_customer, ":", close_value)
```
