```python
#1 Create a dictionary containing only those students whose highest mark is greater than the overall average of the highest marks. 

students = [ 
["Rahul",76,83,79], 
["Neha",94,90,88], 
["Amit",61,67,65], 
["Riya",97,95,99], 
["Jay",81,84,80] 
]

stu_mark = {}
total_mark = 0
total_count = 0

for row in students:
    student = row[0]
    m1 = row[1]
    m2 = row[2]
    m3 = row[3]

    if m1 > m2 and m1 > m3:
        stu_mark[student] = m1
        total_mark += m1
    elif m2 > m1 and m2 > m3:
        stu_mark[student] = m2
        total_mark += m2
    else:
        stu_mark[student] = m3
        total_mark += m3
    total_count += 1
average = total_mark/total_count

new = {}
for student,marks in stu_mark.items():
    if marks > average:
        new[student] = marks

print(new)
```

```python
#2 Count how many products have a price greater than the overall average price. 

products = { 
"Laptop":[78000,7], 
"Mobile":[41000,15], 
"Tablet":[29000,9], 
"TV":[86000,4], 
"Monitor":[53000,10], 
"Speaker":[19000,18] 
}
pro_price = {}
total_price = 0
total_count = 0

for key,value in products.items():
    pro_price[key] = value[0]
    total_price += value[0]
    total_count += 1
average = total_price/total_count

count = 0
for product,price in pro_price.items():
    if price > average:
        count += 1

print(count) 
```

```python
#3 Find the department having the highest average salary below the company average salary. 

employees = [ 
["Rahul","IT",64000], 
["Neha","IT",70000], 
["Amit","HR",44000], 
["Riya","HR",56000], 
["Jay","Sales",66000], 
["Karan","Sales",62000], 
["Pooja","Finance",91000], 
["Raj","Finance",49000] 
]

dep_salary = {}
dep_count = {}
sal_total = 0
emp_count = 0

for row in employees:
    department = row[1]
    salary = row[2]

    if department in dep_salary:
        dep_salary[department] += salary
        dep_count[department] += 1
    else:
        dep_salary[department] = salary
        dep_count[department] = 1
    sal_total += salary
    emp_count += 1
c_average = sal_total/emp_count

new = {}
for dep,sal in dep_salary.items():
    d_average = dep_salary[dep] / dep_count[dep]

    if d_average < c_average:
        new[dep] = sal

print(new)
```

```python
#4 Create a new list containing only those numbers whose difference between first and last digit is odd. 

numbers = [12,23,34,45,56,67,78,89,91,105] 

new = []
for i in numbers:
    txt = str(i)
    dif = max(int(txt[0]),int(txt[-1])) - min(int(txt[0]),int(txt[-1]))

    if dif % 2 != 0:
        new.append(i)
        
print(new)
```

```python
#5 Count how many customers placed at least two but not more than three orders. 

orders = [ 
["Rahul"], 
["Rahul"], 
["Rahul"], 
["Amit"], 
["Amit"], 
["Neha"], 
["Neha"], 
["Jay"], 
["Jay"], 
["Jay"], 
["Jay"], 
["Riya"], 
["Karan"], 
["Karan"] 
]

cus_order = {}

for row in orders:
    customer = row[0]

    if customer in cus_order:
        cus_order[customer] += 1
    else:
        cus_order[customer] = 1

count = 0
for customer,order in cus_order.items():
    if order >= 2 and order <= 3:
        count += 1

print(count)
```

```python
#6 Create a dictionary whose key is department and value is average salary, but include only departments whose average salary is greater than the company average salary. 

employees = [ 
["Rahul","IT",69000], 
["Neha","IT",71000], 
["Amit","HR",46000], 
["Riya","HR",50000], 
["Jay","Sales",65000], 
["Karan","Sales",69000], 
["Pooja","Finance",87000], 
["Raj","Finance",53000] 
]

dep_salary = {}
dep_count = {}
total_sal = 0
total_count = 0

for row in employees:
    department = row[1]
    salary = row[2]

    if department in dep_salary:
        dep_salary[department] += salary
        dep_count[department] += 1
    else:
        dep_salary[department] = salary
        dep_count[department] = 1
    total_sal += salary
    total_count += 1
c_average = total_sal/total_count

new = {}
for department,salary in dep_salary.items():
    d_average = dep_salary[department] / dep_count[department]

    if d_average > c_average:
        new[department] = d_average

print(new)
```

```python
#7 Find the product whose stock value (price × quantity) is closest to the overall average stock value. 

products = { 
"Laptop":[74000,8], 
"Mobile":[39000,16], 
"Tablet":[26000,11], 
"TV":[91000,5], 
"Monitor":[51000,9], 
"Speaker":[17000,20] 
}

o_value = {}
total_value = 0
total_count = 0

for key,value in products.items():
    o_value[key] = value[0]*value[1]
    total_value += value[0]*value[1]
    total_count += 1
average = total_value/total_count

var_price = float("inf")
var_pro = None

for product,value in o_value.items():
    if abs(average - value) < var_price:
        var_price = abs(average - value)
        var_pro = product

print(var_pro,":",var_price)
```

```python
#8 Count how many students scored below the class average in exactly two subjects. 

students = [ 
["Rahul",82,77,89], 
["Neha",92,95,90], 
["Amit",65,70,67], 
["Riya",98,96,99], 
["Jay",79,81,84] 
] 

m1_total = 0
m2_total = 0
m3_total = 0
total_count = 0

for row in students:
    student = row[0]
    m1 = row[1]
    m2 = row[2]
    m3 = row[3]

    m1_total += m1
    m2_total += m2
    m3_total += m3
    total_count += 1

m1_avg = m1_total/total_count
m2_avg = m2_total/total_count
m3_avg = m3_total/total_count

count = 0
for row in students:
    student = row[0]
    m1 = row[1]
    m2 = row[2]
    m3 = row[3]

    below = 0

    if m1 < m1_avg:
        below += 1
    if m2 < m2_avg:
        below += 1
    if m3 < m3_avg:
        below += 1

    if below == 2:
        count += 1

print(count)
```

```python
#9 Create a new list containing only those numbers whose sum of first and last digit is greater than the average sum of all numbers. 

numbers = [14,25,36,47,58,69,71,82,93] 

total = 0
count = 0

for i in numbers:
    txt = str(i)
    add = int(txt[0]) + int(txt[-1])
    total += add
    count += 1
average = total/count

new = []
for i in numbers:
    txt = str(i)
    add = int(txt[0]) + int(txt[-1])
    if add > average:
        new.append(i)

print(new)
```
