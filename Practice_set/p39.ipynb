```python
#1 Create a dictionary containing only those students whose average mark is greater than the class average, but whose lowest mark is below the class lowest-mark average.

students = [ 
["Rahul",81,75,94], 
["Neha",93,89,91], 
["Amit",62,68,71], 
["Riya",98,94,96], 
["Jay",84,79,86] 
]

stu_avg = {}
total_avg = 0
avg_count = 0

for row in students:
    student = row[0]
    m1 = row[1]
    m2 = row[2]
    m3 = row[3]
    stu_avg[student] = (m1+m2+m3)/3

    total_avg += (m1+m2+m3)/3
    avg_count += 1
avg_avg = total_avg/avg_count

stu_low = {}
total_low = 0
total_count = 0

for row in students:
    student = row[0]
    m1 = row[1]
    m2 = row[2]
    m3 = row[3]

    if m1 < m2 and m1 < m3:
        low = m1
    elif m2 < m1 and m2 < m3:
        low = m2
    else:
        low = m3

    stu_low[student] = low
    total_low += low
    total_count += 1

low_avg = total_low / total_count

new = {}

for student, mark in stu_avg.items():
    if mark > avg_avg and stu_low[student] < low_avg:
        new[student] = mark

print(new)
```

```python
#2 Count how many products have stock value (price × quantity) greater than the average stock value but quantity less than the average quantity. 

products = { 
"Laptop":[76000,8], 
"Mobile":[42000,17], 
"Tablet":[31000,7], 
"TV":[93000,5], 
"Monitor":[52000,10], 
"Speaker":[18000,20] 
}

pro_value = {}
total_value = 0
value_count = 0

for key,value in products.items():
    pro_value[key] = value[0]*value[1]
    total_value += value[0]*value[1]
    value_count += 1
value_avg = total_value/value_count

pro_qty = {}
total_qty = 0
qty_count = 0

for key,value in products.items():
    pro_qty[key] = value[1]
    total_qty += value[1]
    qty_count += 1
qty_avg = total_qty/qty_count

count = 0
for product,values in pro_value.items():
    if values > value_avg and pro_qty[product] < qty_avg:
        count += 1

print(count)
```

```python
#3 Find the department having the highest average salary among departments whose total salary is below the company average department total salary. 

employees = [ 
["Rahul","IT",67000], 
["Neha","IT",73000], 
["Amit","HR",45000], 
["Riya","HR",52000], 
["Jay","Sales",69000], 
["Karan","Sales",61000], 
["Pooja","Finance",92000], 
["Raj","Finance",47000] 
]

dep_sal = {}
emp_count = {}
sal_total = 0
sal_count = 0

for row in employees:
    department = row[1]
    salary = row[2]

    if department in dep_sal:
        dep_sal[department] += salary
        emp_count[department] += 1
    else:
        dep_sal[department] = salary
        emp_count[department] = 1
        sal_count += 1
    sal_total += salary
dep_avg = sal_total/sal_count

high_avg = 0
high_dep = None

for key, value in dep_sal.items():
    avg = dep_sal[key] / emp_count[key]

    if value < dep_avg and avg > high_avg:
        high_avg = avg
        high_dep = key

print(high_dep, ":", high_avg)
```

```python
#5 Count how many customers placed more than one order and whose total order value is greater than the average customer total order value. 

orders = [ 
["Rahul",12000], 
["Rahul",14000], 
["Amit",8000], 
["Amit",9000], 
["Neha",17000], 
["Neha",18000], 
["Jay",10000], 
["Jay",12000], 
["Riya",7000], 
["Karan",11000] 
]

cus_value = {}
emp_dep = {}
value_total = 0
cus_count = 0

for row in orders:
    customer = row[0]
    values = row[1]

    if customer in cus_value:
        cus_value[customer] += values
        emp_dep[customer] += 1
    else:
        cus_value[customer] = values
        emp_dep[customer] = 1
        cus_count += 1
    value_total += values
value_avg = value_total/cus_count

count = 0
for customer,values in cus_value.items():
    if values > value_avg and emp_dep[customer] > 1:
        count += 1

print(count)
```

```python
#6 Create a dictionary whose key is department and value is average salary, but include only departments whose average salary is above the company average salary and whose employee count equals the average department employee count.

employees = [ 
["Rahul","IT",70000], 
["Neha","IT",72000], 
["Amit","HR",46000], 
["Riya","HR",50000], 
["Jay","Sales",68000], 
["Karan","Sales",64000], 
["Pooja","Finance",90000], 
["Raj","Finance",52000] 
]

dep_salary = {}
emp_count = {}
sal_total = 0
emp_total = 0
sal_count = 0
dep_count = 0

for row in employees:
    department = row[1]
    salary = row[2]

    if department in dep_salary:
        dep_salary[department] += salary
        emp_count[department] += 1
    else:
        dep_salary[department] = salary
        emp_count[department] = 1
        dep_count += 1
    sal_total += salary
    emp_total += 1
    sal_count += 1
c_average = sal_total/emp_total

new = {}
for department,salary in dep_salary.items():
    d_average = dep_salary[department] / emp_count[department]

    if d_average > c_average:
        new[department] = d_average

print(new)
```

```python
#7 Find the product whose price-per-unit-stock (price ÷ quantity) is farthest from the average price-per-unit-stock.

products = { 
"Laptop":[72000,9], 
"Mobile":[36000,18], 
"Tablet":[27000,6], 
"TV":[90000,5], 
"Monitor":[54000,12], 
"Speaker":[18000,15] 
}

pro_per = {}
total_value = 0
total_count = 0
for key,value in products.items():
    pro_per[key] = value[0] / value[1]
    total_value += value[0] / value[1]
    total_count += 1
average = total_value / total_count

far_value = 0
far_pro = None

for product,values in pro_per.items():
    if abs(average - values) > far_value:
        far_value = abs(average - values)
        far_pro = product

print(far_pro,":",far_value)
```

```python
#9 Create a new list containing only those numbers whose absolute difference between first and last digit is greater than the average difference and whose digit sum is even. 

numbers = [15,24,37,46,59,68,71,82,94] 

dif_total = 0
dif_count = 0

for i in numbers:
    txt = str(i)
    dif = abs(int(txt[0]) - int(txt[-1]))
    dif_total += dif
    dif_count += 1

average = dif_total / dif_count

new = []
for i in numbers:
    txt = str(i)
    dif = abs(int(txt[0]) - int(txt[-1]))
    digit_sum = int(txt[0]) + int(txt[-1])

    if dif > average and digit_sum % 2 == 0:
        new.append(i)

print(new)
```

```python
#10 Find the customer whose average order value is closest to the overall average order value,
# but only among customers whose total order value is above the overall average customer total order value.

orders = [
["Rahul",12000],
["Rahul",13000],
["Amit",9000],
["Amit",10000],
["Neha",18000],
["Neha",17000],
["Jay",14000],
["Jay",12000],
["Riya",8000],
["Riya",9000]
]

cus_total = {}
cus_count = {}

total_order = 0
order_count = 0

for row in orders:
    customer = row[0]
    value = row[1]

    if customer in cus_total:
        cus_total[customer] += value
        cus_count[customer] += 1
    else:
        cus_total[customer] = value
        cus_count[customer] = 1

    total_order += value
    order_count += 1

overall_order_avg = total_order / order_count

total_customer_value = 0
customer_count = 0

for customer, value in cus_total.items():
    total_customer_value += value
    customer_count += 1

customer_total_avg = total_customer_value / customer_count

close_value = float("inf")
close_customer = None

for customer, value in cus_total.items():
    avg_value = value / cus_count[customer]

    if value > customer_total_avg and abs(avg_value - overall_order_avg) < close_value:
        close_value = abs(avg_value - overall_order_avg)
        close_customer = customer

print(close_customer, ":", close_value)
```
