```python
#1 Take your first name, last name, and age as input. 
"""
Print exactly: 
----- Student Profile ----- 
Name : Rahul Sharma 
Age : 21 
Data Type of Age : <class 'int'> 
"""

f_name = input("Enter your First name: ")
l_name = input("Enter your Last name: ")
age = int(input("Enter your age: "))

print("----- Student Profile -----")
print("Name :", f_name, l_name)
print("Age :", age)
print("Data Type of Age :", type(age))
```

```python
#2 Take two numbers as input. 
"""
Print: 
Before Conversion : <class 'str'> <class 'str'> 
After Conversion : <class 'int'> <class 'int'> 
 
Sum : 35 
Use only one print()
"""

num1 = input("Enter a first number: ")
num2 = input("Enter a second number: ")

print(f"Before Conversion: {type(num1)} {type(num2)} \nAfter Conversion: {type(int(num1))} {type(int(num2))} \nSum : {int(num1)+int(num2)}")
```

```python
#3 Take a decimal number as input. 
"""
Print: 
Original Value : 
Integer Value : 
Float Value : 
String Value : 
Boolean Value : 
Also print the datatype of every converted value. 
"""

num = input("Enter a number: ")

print(f"Original Value : {num} - (Datatype {type(num)})")
print(f"Integer Value : {int(num)} - (Datatype {type(int(num))})")
print(f"Float Value : {float(num)} - (Datatype {type(float(num))})")
print(f"String Value : {str(num)} - (Datatype {type(str(num))})")
print(f"Boolean Value : {bool(num)} - (Datatype {type(bool(num))})")
```

```python
#4 Take three integers as input. 
"""
Print: 
Numbers Entered : 
10 20 30 
 
Total : 
Average : 

Use: 
- sep  
- \n 
"""

num1 = int(input("Enter a first number: "))
num2 = int(input("Enter a second number: "))
num3 = int(input("Enter a third number: "))

print("Numbers Entered :\n", num1, num2, num3, sep=" ")
print(f"Total : {num1+num2+num3}")
print(f"Average : {(num1+num2+num3) / 3}")
```

```python
#5 Take your city, state, and country as input. 
"""
Output: 

Address : 
Surat, Gujarat, India 
Use sep=",". 
"""

city = input("Enter city: ") 
state = input("Enter state: ") 
country = input("Enter country: ")

print("Address :")
print(city, state, country, sep=", ")
```

```python
#6 Take one integer. 
"""
Print the following using only one print().

Entered Number : 25 
Square : 625 
Cube : 15625 

Use \n. 
"""

num = int(input("Enter a number: "))

print(f"Entered Number : {num}\nSquare : {num**2}\nCube : {num**3}")
```

```python
#7 Take your name and salary. 
"""
Convert salary into float and print: 

Employee --------- 
Name : Rahul 
Salary : 45000.0 
Salary Type : <class 'float'> 
"""

salary = int(input("Enter salary amount : "))

print(f"Employee {"-"*9}")
print(f"Name : Rahul \nSalary : {float(salary)} \nSalary Type : {type(float(salary))}")
```

```python
#8 Take two words. 
"""
Print them without space on the first line and with "-" between them on the second line. 

Example: 
DataScience 
Data-Science 

Use: 
end  
sep 
"""

word1 = input("Enter first word: ")
word2 = input("Enter second word: ")

print(word1, word2, sep="", end="\n")
print(word1, word2, sep="-")
```

```python
#9 Take an integer as input. 
"""
Convert it into: 
 float  
 string  
 boolean  

Print in this format: 
Integer : 15 
Float : 15.0 
String : 15 
Boolean : True 

Also print the datatype of each. 
"""

num = int(input("Enter a number :"))

print(f"Integer : {num} - (Datatype {type(num)})")
print(f"Float : {float(num)} - (Datatype {type(float(num))})")
print(f"String : {str(num)} - (Datatype {type(str(num))})")
print(f"Boolean : {bool(num)} - (Datatype {type(bool(num))})")
```

```python
#10 
"""
Take: 
 Name  
 Age  
 Height  
 Weight  

Convert: 
 Age → int  
 Height → float  
 Weight → float  

Print: 
----- Health Card ----- 
Name : Rahul 
Age : 21 
Height : 5.9 
Weight : 68.5 
 
Data Types 
Age : <class 'int'> 
Height : <class 'float'> 
Weight : <class 'float'> 
"""

name = input("Enter name :")
age = int(input("Enter age :"))
Height = float(input("Enter Height: "))
Weight = float(input("Enter Weight: "))

print(f"{"-"*5} Health Card {"-"*5}")
print(f"Name : {name}\nAge : {age}\nHeight : {Height}\nWeight : {Weight}\n")
print("Data Types")
print(f"Age : {type(age)}\nHeight : {type(Height)}\nWeight : {type(Weight)}")
```
