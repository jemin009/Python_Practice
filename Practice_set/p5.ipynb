```python
#19
data = { 
    "A":80, 
    "B":95, 
    "C":70, 
    "D":90 
} 

highest = 0
second_highest = 0
highest_key = ""
second_highest_key = ""

for i in data:
    if data[i] > highest:
        second_highest = highest
        second_highest_key = highest_key

        highest = data[i]
        highest_key = i

    elif data[i] > second_highest:
        second_highest = data[i]
        second_highest_key = i

print("Second Highest Key:", second_highest_key)
print("Second Highest Value:", second_highest)
```
