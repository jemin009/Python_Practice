```python
#1 Find the highest number. 

numbers = [15, 42, 8, 91, 33] 

highest_num = 0

for i in numbers:
    if highest_num < i:
        highest_num = i
    
print(highest_num)
```

```python
#2 Find the smallest number. 

numbers = [55, 12, 78, 9, 34] 
 
lowest_num = float("inf")

for i in numbers:
    if lowest_num > i:
        lowest_num = i

print(lowest_num)
```

```python
#3 Count how many even numbers are present. 

numbers = [12, 7, 18, 25, 30, 41, 50] 

count = 0
for i in numbers:
    if i%2 == 0:
        count += 1 

print(count)
```

```python
#4 Print only numbers greater than 50. 

numbers = [23, 67, 45, 89, 12, 54, 90] 

fifty = []

for i in numbers:
    if i>50:
        fifty.append(i)

print(fifty)
```

```python
#5 Calculate the sum of all odd numbers. 

numbers = [11, 20, 15, 42, 9, 18, 7] 
 
sum = 0
for i in numbers:
    if i%2 != 0:
        sum += i

print(sum)
```

```python
#6 Find the second highest number. 

numbers = [45, 78, 12, 99, 65, 87] 

highest = 0
second_highest = 0

for i in numbers:
    if highest < i:
        second_highest = highest
        highest = i

print(second_highest)
```

```python
#7 Print the names whose marks are greater than 80. 

students = { 
    "Rahul": 75, 
    "Amit": 91, 
    "Neha": 84, 
    "Jay": 69 
} 

new = {}

for key,value in students.items():
    if value > 80:
        new[key] = value

print(new)
```

```python
#8 Find the average marks. 

students = { 
    "A": 80, 
    "B": 90, 
    "C": 70, 
    "D": 60 
} 

total = 0

for key,value in students.items():
    total += value

avg = total/len(students)

print(avg)
```

```python
#9 Create a new dictionary containing only students who scored 80 or above. 

students = { 
    "A": 75, 
    "B": 90, 
    "C": 81, 
    "D": 65, 
    "E": 95 
} 

new = {}

for key,value in students.items():
    if value >= 80:
        new[key] = value

print(new)
```

```python
#10 Find the department having the highest total salary. 

employees = [ 
    ["Rahul", "IT", 50000], 
    ["Amit", "HR", 30000], 
    ["Neha", "IT", 45000], 
    ["Jay", "Sales", 35000], 
    ["Riya", "HR", 40000] 
] 

department_salary = {}

for row in employees:
    department = row[1]
    salary = row[2]

    if department in department_salary:
        department_salary[department] += salary
    else:
        department_salary[department] = salary

highest_salary = 0
highest_department = None

for key,value in department_salary.items():
        if value > highest_salary:
             highest_salary = value 
             highest_department = key

print(highest_department,":",highest_salary)
        
```
